```python
class Car:

    def __init__(self, name, color, price, tank_size):
        self.name = name
        self.color = color
        self.price = price
        self.tank_size = tank_size

    def __repr__(self):
        return f"Car {self.name} | Color {self.color} | Price: {self.price}€ | Tank size: {self.tank_size}l"
```

```python
porsche = Car("Porsche Carrera", "green", 50000, 65)
print(porsche)
```

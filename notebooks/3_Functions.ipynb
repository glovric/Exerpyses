### 1. Write a function which returns the sum of all the numbers in a list.

Example:

```python 
[1, 2, 3, 4, 5, 6] --> 21
```

````python
def add_elements(arr):
    total = 0
    for i in arr:
        total += i
    return total

print(add_elements([1, 2, 3, 4, 5, 6]))
````

### 2. Write a function which returns the sum of all odd numbers in a list.

Example:

```python 
[1, 2, 3, 4, 5, 6] --> 9
```

````python
def add_odd_elements(arr):
    total = 0
    for i in arr:
        if i % 2 == 1:
            total += i
    return total

print(add_odd_elements([1, 2, 3, 4, 5, 6]))
````

### 3. Write a function which returns the sum of all elements which are multiples of number 7 in a list.

Example:

```python 
[1, 5, 7, 9, 12, 14, 18, 19, 35, 44, 161] --> 217
```

````python
def add_7_elements(arr):
    total = 0
    for i in arr:
        if i % 7 == 0:
            total += i
    return total

print(add_7_elements([1, 5, 7, 9, 12, 14, 18, 19, 35, 44, 161]))
````

### 4. Write a function which returns the smallest element in a list.

Example:

```python 
[5, -10, 15, -3, 230, 111, 67] --> -10
```

````python
def find_smallest(arr):
    smallest = 999
    for i in arr:
        if i < smallest:
            smallest = i
    return smallest

print(find_smallest([5, -10, 15, -3, 230, 111, 67]))
````

### 5. Write a function which takes a list as an argument and returns a list containing the squared elements.

Example:

```python 
[-3, -2, -1, 0, 1, 2, 3, 4, 5] --> [9, 4, 1, 0, 1, 4, 9, 16, 25]
```

````python
def get_squares(arr):
    result = []
    for i in arr:
        result.append(i**2)    
    return result

print(get_squares([-3, -2, -1, 0, 1, 2, 3, 4, 5]))
````

### 6. Write a function which takes a list of strings as an argument and returns a dictionary which maps strings to string lengths.

Example:

```python 
["mockingbird", "candy", "balloon", "overview", "Arnold"] --> {'mockingbird': 11, 'candy': 5, 'balloon': 7, 'overview': 8, 'Arnold': 6}
```

````python
def get_dict_lengths(arr):
    result = {}
    for i in arr:
        result[i] = len(i)  
    return result

print(get_dict_lengths(["mockingbird", "candy", "balloon", "overview", "Arnold"]))
````

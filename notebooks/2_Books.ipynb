### 1. Write a Book class with the following attributes:

    title
    author
    number of pages
    year

````python
class Book:

    def __init__(self, title, author, num_pages, year):
        self.title = title
        self.author = author
        self.num_pages = num_pages
        self.year = year
````

### 2. Create the Book instances with the following information:

| Author                | Title                         | Number of pages | Year |
|-----------------------|-------------------------------|------------------|------|
| Anonymous             | Beowulf                       | 320              | 975 |
| Niccolò Machiavelli   | The Prince                    | 140              | 1532 |
| William Shakespeare   | Hamlet                        | 160              | 1603 |
| Jane Austen           | Pride and Prejudice          | 432              | 1800 |
| Mary Shelley          | Frankenstein                 | 280              | 1818 |
| J. D. Salinger        | The Catcher in the Rye       | 255              | 1951 |

Let the number of pages and year be in integer format.

````python
book1 = Book("Beowulf", "Anonymous", 320, 975)
book2 = Book("The Prince", "Niccolo Machiavelli", 140, 1532)
book3 = Book("Hamlet", "William Shakespeare", 160, 1603)
book4 = Book("Pride and Prejudice", "Jane Austen", 432, 1800)
book5 = Book("Frankenstein", "Mary Shelley", 280, 1818)
book6 = Book("The Catcher in the Rye", "J. D. Salinger", 255, 1951)
````

### 3. Suppose you have a list of books which represents books available at a library.

### Create a function `oldest_book` which takes a list of books as an argument and it returns the year of the oldest book from the library

Example:

```python 
>>> oldest_book(books)
>>> 975
```

````python
def oldest_book(books):
    oldest = 9999
    for b in books:
        if b.year < oldest:
            oldest = b.year
    return oldest

books = [book1, book2, book3, book4, book5, book6]
print(oldest_book(books))
````

### 4. Create a function `shortest_author` which takes a list of books as an argument and it returns the title of the book whose author's name is the shortest.

Example:

```python 
>>> shortest_author(books)
>>> Beowulf
```

````python
def shortest_author(books):
    shortest_len = 999
    shortest_book = ''
    for b in books:
        author_len = len(b.author)
        if author_len < shortest_len:
            shortest_len = author_len
            shortest_book = b.title
    return shortest_book

print(shortest_author(books))
````

### 5. Create a function `average_pages` which takes a list of books as an argument and prints the average number of all the books.

Example:

```python 
>>> average_pages(books)
>>> The average number of pages is: 264.5
```

````python
def average_pages(books):
    total = 0
    for b in books:
        total += b.num_pages
    average = total / len(books)
    print('The average number of pages is:', average)

average_pages(books)
````

### 6. Create a function `book_century` which takes a list of books as an argument and returns a dictionary in which keys are book names and values are the century in which the book was written. 

Example:

```python 
>>> book_century(books)
>>> {'Beowulf': 10, 'The Prince': 16, 'Hamlet': 17, 'Pride and Prejudice': 18, 'Frankenstein': 19, 'The Catcher in the Rye': 20}
```

````python
def book_century(books):
    dico = {}
    for b in books:
        century = (b.year - 1) // 100 + 1
        dico[b.title] = century
    return dico

book_dico = book_century(books)
print(book_dico)
````
